````python
%%sql
SELECT *
FROM RETAIL_ORDER, ORDER_ITEM;
````

````python
%%sql
SELECT *
FROM RETAIL_ORDER, ORDER_ITEM
WHERE RETAIL_ORDER.OrderNumber = ORDER_ITEM.OrderNumber;
````

````python
%%sql
/* THETA JOIN */
SELECT *
FROM SKU_DATA AS SD1, SKU_DATA AS SD2
WHERE (SD1.SKU - SD2.SKU) BETWEEN 0 AND 1000;
````

````python
%%sql
/* INNER JOIN*/
SELECT *
FROM RETAIL_ORDER INNER JOIN ORDER_ITEM ON
(RETAIL_ORDER.OrderNumber = ORDER_ITEM.OrderNumber);
````

````python
%%sql
/* LEFT OUTER JOIN*/
SELECT *
FROM RETAIL_ORDER LEFT JOIN ORDER_ITEM ON
(RETAIL_ORDER.OrderNumber = ORDER_ITEM.OrderNumber);
````

````python
%%sql
/* RIGHT OUTER JOIN*/
SELECT *
FROM RETAIL_ORDER RIGHT JOIN ORDER_ITEM ON
(RETAIL_ORDER.OrderNumber = ORDER_ITEM.OrderNumber);
````

````python
%%sql
/* CHAINED JOINS*/
SELECT *
FROM RETAIL_ORDER 
JOIN ORDER_ITEM ON (RETAIL_ORDER.OrderNumber = ORDER_ITEM.OrderNumber)
JOIN SKU_DATA ON (SKU_DATA.SKU = ORDER_ITEM.SKU)
````

````python
%%sql
-- SUBQUERIES: AGGREGATES IN THE WHERE CLAUSE
SELECT *
FROM ORDER_ITEM
WHERE ExtendedPrice >
(SELECT AVG(ExtendedPrice) FROM ORDER_ITEM);
````
